## Load dataframe
When a TMT run is quantified with MaxQuant, a file called proteinGroups.txt is generated. Among other things, this file contains the reporter intensities from each of the channels divided by each of the protein groups. Our module load_df reads this data. Data from another program may also be used if they are formatted in the same way and the labels are provided.

In python, running <code>import load_df</code> will allow us to access the module.

```python
import load_df
```

Then we can load data from the file using <code>load_df.load_dataset()</code> <br>If no filename is provided, this will open a dialog box where you can select one. Here we know the filename and so provide the path like this:

```python
all_data = load_df.load_dataset("data\proteinGroups_PNNL_TMT_SC_Chip1.txt")
```

This file was generated by MaxQuant. Three TMT runs, Chip1 A, Chip 1 B, and Chip 1 C were analyzed together. We can use these experiment names to get the channels from just one run. We use the same call but add <code>experiment='Chip1 A'</code>. This is what we'll do here.

```python
data = load_df.load_dataset("data\proteinGroups_PNNL_TMT_SC_Chip1.txt",
                           experiment='Chip1 A')
```

Columns may also be chosen explicitly by passing a list to <code>usecols</code>. Run <code>print_columns("your_file.txt")</code> to see a list of the columns in your file. 

<code>load_dataset</code> returned a pandas dataframe. Now we have the intensities from Chip 1 row A. We can use the function <code>head</code> to see the first few entries.
